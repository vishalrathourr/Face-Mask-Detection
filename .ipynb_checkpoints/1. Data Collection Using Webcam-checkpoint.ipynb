```python
haar_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
haar_data.detectMultiScale(img)
```

```python
haar_data.detectMultiScale(img2)
```

```python
# cv2.rectangle(img,(x,y),(w,h),(b,g,r),border_thickness)
```

```python
while True:
    faces = haar_data.detectMultiScale(img)
    for x , y, w, h in faces:
        cv2.rectangle(img, (x,y), (x+w, y+h), (255,0, 255),4)        
    cv2.imshow('result',img)
    
    # 27 = ASCII code of Escape
    if cv2.waitKey(2) == 27:
        break
    
cv2.destroyAllWindows( )
```

```python
while True:
    faces = haar_data.detectMultiScale(img2)
    for x , y, w, h in faces:
        cv2.rectangle(img2, (x,y), (x+w, y+h), (255,0, 255),4)        
    cv2.imshow('result2',img2)
    # 27 = ASCII code of Escape
    if cv2.waitKey(2) == 27:
        break
    
cv2.destroyAllWindows( )
```

```python
capture = cv2.VideoCapture(0)
# 0:Default camera, 1: External camera 
# for collecting data from video pass the path of video "video.mp4"
data = []
while True:
    flag, img = capture.read()
    if flag:
        faces = haar_data.detectMultiScale(img)
        for x , y, w, h in faces:
            cv2.rectangle(img, (x,y), (x+w, y+h), (255,0, 255),4)
            face = img[y:y+h, x:x+w, :]
            face = cv2.resize(face, (50,50))
            print(len(data))
            if len(data)< 6000:
                data.append(face)
        cv2.imshow('result',img)
        # 27 = ASCII code of Escape
        if cv2.waitKey(2) == 27 or len(data) >= 400:
            break
capture.release()
cv2.destroyAllWindows( )
```

```python
np.save('without_mask.npy', data)
```

```python
np.save('with_mask.npy', data)
```
